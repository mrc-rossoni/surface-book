```python
import numpy as np
import matplotlib.pyplot as plt
from code.curves import bspline_eval
```

```python
P = np.array([
    [0.0, 0.0],
    [0.2, 1.0],
    [0.6, 1.2],
    [1.0, 0.0],
    [1.2, -0.5],
    [1.5, 0.2]
])

p = 3  # cubic B-spline

# Open-uniform knot vector
U = np.array([0,0,0,0, 1,2,3, 4,4,4,4], dtype=float)
```

```python
u_vals = np.linspace(U[p], U[-p-1], 250)

C = np.array([bspline_eval(P, U, p, u) for u in u_vals])

plt.figure()
plt.plot(P[:,0], P[:,1], "--o", label="Control polygon")
plt.plot(C[:,0], C[:,1], label="B-spline curve")
plt.axis("equal")
plt.grid(True)
plt.legend()
plt.show()
```
