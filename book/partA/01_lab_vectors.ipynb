```python
import numpy as np
import matplotlib.pyplot as plt

P0 = np.array([0.0, 0.0])
P1 = np.array([1.0, 1.0])

u = np.linspace(0, 1, 50)
P = (1-u)[:,None]*P0 + u[:,None]*P1

plt.figure()
plt.plot(P[:,0], P[:,1], label="Interpolated points")
plt.scatter([P0[0], P1[0]], [P0[1], P1[1]], label="Endpoints")
plt.axis("equal")
plt.grid(True)
plt.legend()
plt.show()
```

```python
theta = np.deg2rad(30)
R = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])
t = np.array([0.5, -0.2])

P_trans = (R @ P.T).T + t

plt.figure()
plt.plot(P[:,0], P[:,1], label="Original")
plt.plot(P_trans[:,0], P_trans[:,1], label="Transformed")
plt.axis("equal")
plt.grid(True)
plt.legend()
plt.show()
```
